```python
import pandas as pd
import numpy as np

import tweepy

from tqdm import tqdm
```

```python
auth = tweepy.OAuthHandler('', '')
auth.set_access_token('', '')

api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
df = pd.DataFrame(columns=[
    'username',
    'description',
    'location',
    'following',
    'followers',
    'totaltweets',
    'retweetcount',
    'favoritecount',
    'text',
    'hashtags',
    'created_at',
    'tweet_id',
])
msgs = []
msg =[]
```

```python
if len(df['tweet_id']) > 0:
    cur = tweepy.Cursor(api.search_tweets,
                                   "#NFT", lang="en",
                                   tweet_mode='extended', result_type='popular', max_id=max(df['tweet_id']))
else: 
    cur = tweepy.Cursor(api.search_tweets,
                                   "#NFT", lang="en",
                                   tweet_mode='extended', result_type='popular')

if len(df['tweet_id']) > 0:
    cur = tweepy.Cursor(api.search_tweets,
                                   "#NFT", lang="en",
                                   tweet_mode='extended', result_type='recent', max_id=max(df['tweet_id']))
else: 
    cur = tweepy.Cursor(api.search_tweets,
                                   "#NFT", lang="en",
                                   tweet_mode='extended', result_type='recent')

max_count = 10000
for tweet in tqdm(cur.items(max_count), total=max_count):
    username = tweet.user.screen_name
    description = tweet.user.description
    location = tweet.user.location
    following = tweet.user.friends_count
    followers = tweet.user.followers_count
    totaltweets = tweet.user.statuses_count
    
    
    
    retweetcount = tweet.retweet_count
    favoritecount = tweet.favorite_count
    hashtags = tweet.entities['hashtags']
    created_at = tweet.created_at
    tweet_id = tweet.id

    # Retweets can be distinguished by
    # a retweeted_status attribute,
    # in case it is an invalid reference,
    # except block will be executed
    try:
            text = tweet.retweeted_status.full_text
    except AttributeError:
            text = tweet.full_text
    hashtext = list()
    for j in range(0, len(hashtags)):
            hashtext.append(hashtags[j]['text'])
    
    # Here we are appending all the
    # extracted information in the DataFrame
    ith_tweet = [username, description,
                 location, following,
                 followers, totaltweets,
                 favoritecount, retweetcount, text, hashtext, created_at, tweet_id]
    df.loc[len(df)] = ith_tweet

df
```
